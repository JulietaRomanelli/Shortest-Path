De ese data frame, me quedo unicamente con las que corresponden a mi padron, que son las de la fila del 3.

```python
weight = df.iloc[3]
weight
```

```python
import networkx as nx
import matplotlib.pyplot as plt
g = nx.Graph()

g.add_edge(1,2,weight=weight[0])
g.add_edge(1,3,weight=weight[1])
g.add_edge(1,4,weight=weight[2])
g.add_edge(2,5,weight=weight[3])
g.add_edge(2,6,weight=weight[4])
g.add_edge(3,2,weight=weight[5])
g.add_edge(3,4,weight=weight[6])
g.add_edge(3,6,weight=weight[7])
g.add_edge(4,6,weight=weight[8])
g.add_edge(4,7,weight=weight[9])
g.add_edge(5,8,weight=weight[10])
g.add_edge(6,5,weight=weight[11])
g.add_edge(6,7,weight=weight[12])
g.add_edge(6,8,weight=weight[13])
g.add_edge(7,8,weight=weight[14])


nodos = nx.shortest_path(g,source=1,target=8,method='dijkstra', weight = 'weight')
print(f'El camino mas corto es aquel que pasa por los nodos {nodos}')
distance = nx.shortest_path_length(g,source=1,target=8,method='dijkstra', weight = 'weight')
print(f"La distacia recorrida es de {distance}")
```
